```python
import pandas as pd
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report

X_train = pd.read_csv(r'C:\Users\Divyansh\Desktop\har_project\data\processed\X_train.csv')
X_test = pd.read_csv(r'C:\Users\Divyansh\Desktop\har_project\data\processed\X_test.csv')
y_train = pd.read_csv(r'C:\Users\Divyansh\Desktop\har_project\data\processed\y_train.csv').squeeze() - 1
y_test = pd.read_csv(r'C:\Users\Divyansh\Desktop\har_project\data\processed\y_test.csv').squeeze() - 1

xgb = XGBClassifier(random_state=42)
xgb.fit(X_train, y_train)
y_pred = xgb.predict(X_test)

print(f"Accuracy: {accuracy_score(y_test, y_pred):.4f}")
print(classification_report(y_test, y_pred))
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
import pandas as pd

X_train = pd.read_csv(r'C:\Users\Divyansh\Desktop\har_project\data\processed\X_train.csv')
X_test = pd.read_csv(r'C:\Users\Divyansh\Desktop\har_project\data\processed\X_test.csv')
y_train = pd.read_csv(r'C:\Users\Divyansh\Desktop\har_project\data\processed\y_train.csv').squeeze()
y_test = pd.read_csv(r'C:\Users\Divyansh\Desktop\har_project\data\processed\y_test.csv').squeeze()

param_grid = {
    'C': [0.1, 1, 10],
    'gamma': [0.001, 0.01, 0.1]
}

svm = SVC(kernel='rbf')
grid_search = GridSearchCV(svm, param_grid, cv=5, scoring='accuracy', verbose=2, n_jobs=-1)
grid_search.fit(X_train, y_train)

print(f"Best parameters: {grid_search.best_params_}")
print(f"Best CV accuracy: {grid_search.best_score_:.4f}")

best_svm = grid_search.best_estimator_
y_pred = best_svm.predict(X_test)
print(f"Test accuracy: {accuracy_score(y_test, y_pred):.4f}")
print(classification_report(y_test, y_pred))
```
